```python
def plot_throughput(consumer,t):

    x=[]
    for i in range(len(consumer)):
        x.append(i)
    labels=x
    producer=[]
    sum=0
    for i in range(len(t)):
        sum+=t[i]
        if(i==0):
            continue
        if(i%10==9):
            producer.append(sum/10)
            sum=0
    producer.append(sum/10)
    # Plotting x-axis and y-axis
    sum=0
    for i in consumer:
        sum+=i
    print("Consumer avg throughput ",sum/len(consumer))

    sum=0
    for i in producer:
        sum+=i
    print("Producer avg throughput ",sum/len(producer))
    plt.plot(x, consumer)
    plt.plot(x,producer)
    # naming of x-axis and y-axis
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    
    # naming the title of the plot
    plt.title("Set X-Axis Values in Matplotlib")
    
    # setting x-axis values
    plt.xticks(x, labels)
    
    plt.show()
```

```python
def plot_latency(y):
    x=[]
    for i in range(len(y)):
        x.append(i)
    labels=x
    sum=0
    for i in y:
        sum+=i
    print("Latency ",sum)
    # Plotting x-axis and y-axis
    plt.plot(x, y)
    
    # naming of x-axis and y-axis
    plt.xlabel("X-Axis")
    plt.ylabel("Y-Axis")
    
    # naming the title of the plot
    plt.title("Set X-Axis Values in Matplotlib")
    
    # setting x-axis values
    plt.xticks(x, labels)
    
    plt.show()
```
